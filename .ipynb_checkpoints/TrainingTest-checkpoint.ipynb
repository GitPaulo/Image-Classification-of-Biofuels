```python
import numpy
import tensorflow as tf
import h5py
import tensorflow.keras
from tensorflow.keras.preprocessing.image import *
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
def preproc(image):
    return image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))
```

```python
model = VGG16(include_top=False)
model.compile(loss = "categorical_crossentropy",optimizer = tensorflow.keras.optimizers.SGD(lr=0.0001, momentum=0.9))
for layer in model.layers:
   layer.trainable = False
dataGenerator = ImageDataGenerator(preprocessing_function=preproc)
trainingdata = dataGenerator.flow_from_directory(class_mode = None,directory="M:\Year4\Dataset-20191113T110123Z-001\Dataset",target_size=(224,224))
testdata = dataGenerator.flow_from_directory(class_mode = None,directory="M:\Year4\Dataset-20191113T110123Z-001\Dataset",target_size=(224,224))
```

```python
checkpoint = ModelCheckpoint("vgg16_1.h5", monitor='val_acc', verbose=1, save_best_only=True, save_weights_only=False, mode='auto', period=1)
early = EarlyStopping(monitor='val_acc', min_delta=0, patience=20, verbose=1, mode='auto')
model.fit_generator(steps_per_epoch=100,generator=trainingdata, validation_data= testdata, validation_steps=10,epochs=100,callbacks=[checkpoint,early])
```
